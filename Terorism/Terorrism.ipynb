```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/globalterrorismdb.csv",encoding='latin1')
copy_df=df
```

```python
copy_df.drop(['approxdate','resolution','location','summary','alternative','alternative_txt','attacktype2','attacktype2_txt','attacktype3','attacktype3_txt','corp2','target2','natlty2','natlty2_txt','targtype3','targtype3_txt','targsubtype3','targsubtype3_txt','corp3','target3','natlty3','natlty3_txt','targtype2','targtype2_txt','targsubtype2','targsubtype2_txt',],axis=1,inplace=True)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(copy_df.corr())
```

```python
copy_df.nkillus.plot(kind = 'line', color = 'red', label = 'The Number of Total Confirmed Fatalities for US', linewidth = 2, alpha = 0.8, grid = True, 
                 linestyle = ':', figsize = (10,15), fontsize=15)
copy_df.nwoundus.plot(color = "green", label = 'The Number of Confirmed Non-Fatal Injuries for US', linewidth = 2, alpha = 0.8, grid = True, 
                 linestyle = '-.', figsize = (10,15), fontsize=15)

plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('Database Index', fontsize=10)              # label = name of label
plt.ylabel('Number of Dead or Injuries', fontsize=15)

plt.title('Confirmed Fatalities & Non-Fatal Injuries for US')            #plot title
plt.show()
```

```python
copy_df.plot(kind = 'scatter', x = 'nkill', y = 'nwound', alpha = 0.5, color = 'red', figsize = (10,10), fontsize=15)
plt.xlabel('Kill', fontsize=15)
plt.ylabel('Wound', fontsize=15)
plt.title('Kill - Wound Scatter Plot')
plt.show()
```

```python
print(copy_df['region_txt'].unique())
```

```python
middleEastData = copy_df[copy_df['region_txt'] == 'Middle East & North Africa']
middleEastData.iyear.plot(kind = 'hist', bins = 30, figsize = (10,10), color = 'red', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Frequency of Middle East & North Africa Terrorism Actions by Years')
plt.show()
```

```python
CACData = copy_df[copy_df['region_txt'] == 'Central America & Caribbean']
CACData.iyear.plot(kind = 'hist', bins = 30, figsize = (10,10), color = 'blue', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Frequency of Central America & Caribbean Terrorism Actions by Years')
plt.show()
```

```python
EasternEuropeData = copy_df[copy_df['region_txt'] == 'Eastern Europe']
EasternEuropeData.iyear.plot(kind = 'hist', bins = 30, figsize = (10,10), color = 'orange', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Frequency of Eastern Europe Terrorism Actions by Years')
plt.show()
```

```python
killData = copy_df['nkill']
WoundData = copy_df['nwound']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
print('Number of people Wounded by terror attack:', int(sum(WoundData.dropna())))# drop the NaN values
```

```python
attack_data=copy_df.loc[:,'attacktype1':'attacktype1_txt']
type_kill_data=pd.concat([attack_data,killData],axis=1)
```

```python
typeKillFormatData = type_kill_data.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
typeKillFormatData
```
